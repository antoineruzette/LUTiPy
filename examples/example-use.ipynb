```python
from lutipy.core import LUTiPy
import lutipy.data as data
```

```python
image = data.cells()

# Create a LUTiPy object
lutipy = LUTiPy(layout="horizontal", pixel_size="0.31 nm", scale_length=0.25)
lutipy.process_image(image)
```

```python
image = data.nori()

# Create a LUTiPy object
lutipy = LUTiPy(layout="grid", pixel_size="3 um", scalebar=True)
lutipy.process_image(image)
```

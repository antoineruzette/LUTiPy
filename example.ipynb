```python
import os

import lutipy.core as core
import tifffile as tiff

import matplotlib.pyplot as plt
```

```python
data_dir = "/Users/ranit/IAC/Project/Will Trim/_DATA/ALL"
image = tiff.imread(f"{data_dir}/Fused_S5.tif")
image = core.convert_to_channel_last(image)
```

```python
img_8bit = core.convert_to_8bit(image[200:500,200:500,:3])

plt.imshow(img_8bit)
plt.axis('off')
```

```python
num_channels = img_8bit.shape[-1]
luts = core.find_n_complementary_colors(rgb=(255,0,255), n=num_channels)
image_lut = core.apply_complementary_luts(img_8bit, luts)
```

```python
plt.imshow(image_lut)
plt.axis('off')
```

```python
core.create_panel(img_8bit, luts, ["Protein", "Lipid", "Endomucin"], image_lut, )
```
